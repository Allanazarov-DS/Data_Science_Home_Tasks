```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Jahon mamlakatlarining YAIM ni (Yali ichki mahsulot) tahlil qilamiz

```python
gdp = pd.read_csv("GPD by Country.csv")
e12 = 10**12
gdp = gdp.sort_values(by = 'GDP', ascending = False) # Davlatlarning 5yillik (2017-2021) gdplarini sortlab oldim
gdp['GDP'] = [f"{x / e12:.2f}trln" for x in gdp['GDP']] # - Tushunishga oson bo'lishi uchun GDPni trillionga o'tkazib oldim 
gdp.reset_index(drop=True, inplace=True) # Betartib indekslarni tartiblab oldim huddi reyting kabi
gdp.head(20)
```

### 2017-2021da 5ta davlat iqtisodiyotni ketinma-ketin top 5da bo'lgan ekan ularning grafiklarini chiqaramiz

```python
top_4_countries = gdp.head(20)
top_4_countries['GDPint'] = [int(float(x[:-4])) for x in top_4_countries['GDP']] # - 'GDP' ning int tipidagi qatori kerak boldi shunga qoshdim
sns.lineplot(x = 'Year', y = 'GDPint', data = top_4_countries, hue = 'Country Name', style = 'Country Name', marker = 'o')
plt.legend(loc='upper left', bbox_to_anchor=(0.685, 0.485)) # - labelni boshqarish uchun ishlatdim
plt.show()
```

#### Grafikdan shuni ko'rishimiz mumkinki Xitoy va Amerika Iqtisodiyoti stabil o'sish ko'rsatgan Yaponiya iqtisodiyoti uzoq stabillikdan keyin pasaygan bo'lsa Germaniya iqtisodiyoti oshgan

#### Kishi boshiga to'g'ri keladigan yalpi daromatning (GDP per capita) shu davlatlar uchun 2017-2021da grafigini ko'ramiz (Bu davlatlar yalpi daromadi kishi boshiga hisoblanganda top 5 emas!!!) 

```python
top_4_countries = gdp.head(20)
sns.lineplot(x = 'Year', y = 'GDP per Capita', data = top_4_countries, hue = 'Country Name', style = 'Country Name', marker = 'o')
plt.legend(loc='upper left', bbox_to_anchor=(0.685, 0.485)) # - labelni boshqarish uchun ishlatdim

plt.title("""Gross per capita of the top 4 GDP giants 
income growth graph""")

plt.show()
```

### Endi qaysi fudbolchi ko'roq maydonda to'p surganini aniqlab olamiz

```python
pd.options.mode.copy_on_write = True
player_data = pd.read_csv('RonaldovsMessi.csv')
player_data['Minute'] = player_data['Minute'].str.split('+').apply(lambda x: sum(map(int, x))) # Daqiqalarni (Minute) stringdan intga o'tkazdim 
player_data.head(20)
```

```python
sns.set_palette('colorblind')
sns.barplot(x = 'Player', y = 'Minute', data = player_data, hue = 'Player')
plt.title("Ronaldo va Messning maydonda o'tkazgan o'rtacha vaqti")
plt.ylabel("Minut")
plt.xlabel("O'yinchilar")
plt.show()
```

```python
player_data[player_data['Player'] == 'Cristiano Ronaldo']['Minute'].mean() # - Ronaldo maydonda o'ynagan o'rtacha vaqt
```

```python
player_data[player_data['Player'] == 'Lionel Messi']['Minute'].mean() # - Messi maydonda o'rtacha o'ynagan holat
```

#### Grafiklardan shuni ko'rishimiz mumkinki Messing o'rtacha maydonda o'tkazgan vaqti Ronaldonikidan 1daqiqaga ko'proq ekan

### Boxplotlarga murojat qilib ko'ramiz

```python
sns.set_palette('colorblind')
sns.boxplot(x = 'Player', y = 'Minute', data = player_data, hue = 'Player', notch = True)

plt.show()
```
