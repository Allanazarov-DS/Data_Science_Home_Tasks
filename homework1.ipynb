```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("diabetes_clean.csv")
df.head()
```

### Ustunlar qanchalik bir biriga bog'liqligini bilish uchun korelatsiyasi ko'ramiz

```python
sns.heatmap(df.corr(), annot=True)
plt.show()
```

#### Jadvalda bir biriga bog'liq bo'lgan ustunlarni ko'rishimiz mumkin

```python
model = LinearRegression()
X = df[["diastolic"]]
y = df["triceps"].values
model.fit(X, y)
y_p = model.predict(X)
sns.scatterplot(data = df, x= "diastolic", y="triceps")
plt.plot(X, y_p, c ="k")
plt.show()
```

```python
X = df[["pregnancies"]]
y = df["age"].values
model.fit(X, y)
y_p = model.predict(X)
sns.scatterplot(data = df, x= "pregnancies", y="age")
plt.plot(X, y_p, c ="k")
plt.show()
```
