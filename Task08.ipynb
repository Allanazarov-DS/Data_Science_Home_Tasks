```python
#### Qaysi davlat taomlarining daromadi yuqori ekanligini boxplot bilan ko'ramiz
```

```python
sns.set_style('darkgrid')
sns.catplot(y = 'Revenue', x = 'Cuisine', data = df_restaurant, kind = 'box', hue = 'Cuisine')
plt.title("Resturants")
plt.show()
```

#### Har tomonlama Yapon oshxonasi barcha oshxonalarga qaraganda ko'proq daromad topgan umumiy holatda

#### Endi tahlilimiz adohlatni bo'lishi uchun top har bir davlat oshxonasining top 10talikdaiglarining daromatlarini solishtiramiz

```python
Japan = df_restaurant[df_restaurant['Cuisine'] == 'Japanese'][['Cuisine','Revenue']]
Japan_top10 = Japan.sort_values('Revenue', ascending=False).iloc[:10]
Mexico = df_restaurant[df_restaurant['Cuisine'] == 'Mexican'][['Cuisine','Revenue']]
Mexico_top10 = Mexico.sort_values('Revenue', ascending=False).iloc[:10]
Italy = df_restaurant[df_restaurant['Cuisine'] == 'Italian'][['Cuisine','Revenue']]
Italy_top10 = Italy.sort_values('Revenue', ascending=False).iloc[:10]
France = df_restaurant[df_restaurant['Cuisine'] == 'French'][['Cuisine','Revenue']]
France_top10 = France.sort_values('Revenue', ascending=False).iloc[:10]
India = df_restaurant[df_restaurant['Cuisine'] == 'Indian'][['Cuisine','Revenue']]
India_top10 = India.sort_values('Revenue', ascending=False).iloc[:10]
America = df_restaurant[df_restaurant['Cuisine'] == 'American'][['Cuisine','Revenue']]
America_top10 = America.sort_values('Revenue', ascending=False).iloc[:10]
```

```python
top_10 = pd.concat([Japan_top10, Mexico_top10, Italy_top10, France_top10, India_top10, America_top10])
top_10
```

#### Har bir millatning top 10 daromad ko'radigan oshxonalarini birlashtirib oldim

#### Endi ularni daromadi bo'yicha bar bilan solishtiramiz

```python
sns.catplot(x = 'Cuisine', y = 'Revenue', data = top_10, kind = 'bar', hue = 'Cuisine')
plt.title('Revenue of the top 10 cuisines of each country')
plt.show()
```
