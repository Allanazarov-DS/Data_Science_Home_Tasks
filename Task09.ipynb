### Stripplot orqali turli hududlarning marketinnga ajratadigan mablag'ini ko'ramiz

```python
sns.stripplot(data = df2, y = 'Location', x = "Marketing Budget", jitter= True, hue = 'Marketing Budget', palette = 'YlOrRd')
plt.title("""3ta hududda restaranlarning marketing 
    uchun ajratilgan mablag'lari""")
plt.show()
```
