### Bar chart bilan ishlaymiz

```python
medals = pd.read_csv('medals_by_country_2016.csv', index_col=0)
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'], color = 'green')
plt.show()
# medals.index
```

### Rioda Janero 2016 olimpiyada musbaqalarida davlatlarning ko'rsatgichlari bar chartda ko'rsatildi lekin davlat nomlarini ko'rishning imkoni bolmayapti bunda bizga rotation yordam beradi

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'], color = 'gold')
ax.set_xticklabels(medals.index, rotation = 45)
ax.set_ylabel("Number of medals")
plt.show()
```

### Natija ancha tariblangan holda chiqdi 

### Endigi navbatda kumush medal olganlarni ham oltin medal olganlarning pastgi qismidan chiqaramiz

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'], color = 'gold')
ax.bar(medals.index, medals['Silver'], color = 'silver', bottom = 'Gold')
ax.set_xticklabels(medals.index, rotation = 45)
ax.set_ylabel("Number of medals")
plt.show()
```

### Endi bronza olganlarni ham qoshib olamiz, hamma medal uchun o'z rangini beryapman

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'], color = 'gold')
ax.bar(medals.index, medals['Silver'], color = 'silver', bottom = medals['Gold'])
ax.bar(medals.index, medals['Bronze'], color = 'brown', bottom = medals['Silver'] + medals['Gold'])
ax.set_xticklabels(medals.index, rotation = 45)
ax.set_ylabel("Number of medals")
plt.show()
```

### Natijada bar chart shaklida medallar haqida ma'lumot chiqarib berdi

### subplot qilib ham chiqarib ko'rdim
### fig.text(0.04, 0.5, 'Number of Medals', va='center', rotation='vertical') bu kodni chatgptdan oldim

```python
fig, ax = plt.subplots(3, 1, sharey=True, sharex=True)
ax[0].bar(medals.index, medals['Gold'], color = 'gold')
ax[0].set_xticklabels(medals.index, rotation = 45)
ax[1].bar(medals.index, medals['Silver'], color = 'silver')
ax[1].set_xticklabels(medals.index, rotation = 45)
ax[2].bar(medals.index, medals['Bronze'], color = 'brown')
ax[2].set_xticklabels(medals.index, rotation = 45)
fig.text(0.04, 0.5, 'Number of Medals', va='center', rotation='vertical')
plt.show()
```

### Alohida alohida bar chart qilib chiqarib oldim

### Endi legend qo;shaman

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'], color = 'gold', label = 'Silver')
ax.bar(medals.index, medals['Silver'], color = 'silver', bottom = medals['Gold'], label = 'Gold')
ax.bar(medals.index, medals['Bronze'], color = 'brown', bottom = medals['Silver'] + medals['Gold'], label = 'Bronze')
ax.set_xticklabels(medals.index, rotation = 45)
ax.set_ylabel("Number of medals")
ax.legend()
plt.show()
```

### Rowing bilan shug'ullanadigan erkaklarni alohida dataframega, Gymnastics bilan shug'ullanadigan erkaklarni alohida dataframega o'tkazib olmoqchiman

```python
mens_rowing = pd.read_csv('summer2016.csv').query('Sport == "Rowing" and Sex == "M"')
mens_gymnastics = pd.read_csv('summer2016.csv').query('Sport == "Gymnastics" and Sex == "M"')
```

### query() metodi bilan yozilgan kod ancha tushunarli bo'lgani uchun menga query bilan yozish koproq yoqdi, lekin bunaqa usuli ham bor degan manoda buni ham yozib ketdim har ehtimolga qarshi

### Eshkak eshuvchilar va Gimnast erkaklarning o'rtacha bo'yini bar chartda chiqarish uchun kod yozib kormoqchimana

```python
fig, ax = plt.subplots()
ax.bar('Rowing', mens_rowing['Height'].mean())
ax.bar('Gymnastics', mens_gymnastics['Height'].mean())
ax.set_xlabel("Height (cm)")
plt.show()
```

### Bart chartni chizib oldik. Bundan shu narsa ma'lumki Rowing (Eshkak eshish) sporchilarining o'rtacha bo'yi gimnastlarnikidagan ancha baland ekan

### Histogrammalar

```python
fig, ax = plt.subplots()
ax.hist(mens_rowing['Height'])
ax.hist(mens_gymnastics['Height'])
ax.set_xlabel("Height in cm")
ax.set_ylabel("# of obsevartions")
plt.show()
```

### Legendlar ham qoshib olamiz

```python
fig, ax = plt.subplots()
ax.hist(mens_rowing['Height'], label = 'Rowing')
ax.hist(mens_gymnastics['Height'], label = 'Gymnastics')
ax.set_xlabel("Height in cm")
ax.set_ylabel("# of obsevartions")
ax.legend()
plt.show()
```

### Histogrammaga legendlar qoshib oldim

### bins qo'shib kormoqchiman

```python
fig, ax = plt.subplots()
ax.hist(mens_rowing['Height'], label = 'Rowing', bins = 15)
ax.hist(mens_gymnastics['Height'], label = 'Gymnastics', bins = 15)
ax.set_xlabel("Height in cm")
ax.set_ylabel("# of obsevartions")
ax.legend()
plt.show()
```

### Bins ham qo'shib ko'rdim juda foydalanish uchun qular ekan

### Endi bins ga boundrieslar ham qo'yib chiqaramiz

```python
fig, ax = plt.subplots()
ax.hist(mens_rowing['Height'], label = 'Rowing', bins = [150, 160, 170, 180, 190, 200, 210, 220])
ax.hist(mens_gymnastics['Height'], label = 'Gymnastics', bins = [150, 160, 170, 180, 190, 200, 210, 220])
ax.set_xlabel("Height in cm")
ax.set_ylabel("# of obsevartions")
ax.legend()
plt.show()
```

### Boundries ni qo'lbola to'g'lab oldim ya'ni 210 gacha bo'lgan sportchilar bor ekan maximal minal bo'y 150 ekan sportchilar orasida

### alphani ham ishlatib ko'rmoqchiman

```python
fig, ax = plt.subplots()
ax.hist(mens_rowing['Height'], label = 'Rowing', bins = [150, 160, 170, 180, 190, 200, 210], alpha = 0.5)
ax.hist(mens_gymnastics['Height'], label = 'Gymnastics', bins = [150, 160, 170, 180, 190, 200, 210], alpha = 0.5)
ax.set_xlabel("Height in cm")
ax.set_ylabel("# of obsevartions")
ax.legend()
plt.show()
```

### alpha() metodi yordamida ustunlarni shaffof qilib oldik va shuni korishimiz mumkinki eshkash eshish sportida minimal bo'y 160sm gimnastlarda esa maximal bo'y 190sm ekan


### histogrammalarni rangsiz faqat chiziqlari bilan qoldiramiz

```python
fig, ax = plt.subplots()
ax.hist(mens_rowing['Height'], label = 'Rowing', bins = [150, 160, 170, 180, 190, 200, 210], histtype='step')
ax.hist(mens_gymnastics['Height'], label = 'Gymnastics', bins = [150, 160, 170, 180, 190, 200, 210], histtype='step')
ax.set_xlabel("Height in cm")
ax.set_ylabel("# of obsevartions")
ax.legend()
plt.show()
```

### alphadan ham yaxshiroq visulizatiya berar ekan 

### Error bardan foydalanib ko'ramiz

```python
fig, ax = plt.subplots()
ax.bar("Rowing", mens_rowing['Height'].mean(), yerr = mens_rowing['Height'].std(), label = 'Rowing')
ax.bar("Gymnastics", mens_gymnastics['Height'].mean(), yerr = mens_gymnastics['Height'].std(), label = 'Gymnastics')
ax.set_ylabel("Height (cm)")
ax.legend()
plt.show()
```

### Standart devisionni ko'rsatadigan error barlarni tuzib oldim

```python
fig, ax = plt.subplots()
ax.errorbar(seattle_weather["DATE"], seattle_weather['MLY-TAVG-NORMAL'], yerr = seattle_weather['MLY-TAVG-STDDEV'])
ax.errorbar(austin_weather["DATE"], austin_weather['MLY-TAVG-NORMAL'], yerr = austin_weather['MLY-TAVG-STDDEV'])
ax.set_ylabel("Temperature (Fahrenheit)")
plt.show()
```

### Boxplotni qoshib koramiz


```python
fig, ax = plt.subplots()
ax.boxplot([mens_rowing['Height'], mens_gymnastics['Height']])
ax.set_xticklabels(['Rowing', 'Gymnastics'])

plt.show()
```
