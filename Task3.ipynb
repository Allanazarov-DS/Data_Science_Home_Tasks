### Pivot_table() dan foydalanamiz

```python
data = {
    'A': ['foo', 'foo', 'foo', 'bar', 'bar', 'bar'],
    'B': ['one', 'one', 'two', 'two', 'one', 'one'],
    'C': [1, 3, 2, 5, 4, 6],
    'D': [10, 20, 30, 40, 50, 60]
}

df_pivot = pd.DataFrame(data)
df_pivot
```

```python
df_pivot.pivot_table(values='D', index=['A', 'B'], aggfunc=['sum', 'mean'])
```
